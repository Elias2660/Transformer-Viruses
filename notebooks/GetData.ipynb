#### Accessing the Data

```python
# inporting the required libraries
import numpy as np
import pandas as pd
import difflib as df
```

##### Genome Data

```python
genome_data = pd.DataFrame(
    columns=[
        "Locus",
        "Position/Length Indicator?",
        "Virus Name",
        "Genome",
        "Estimated Transmission Level",
        "temp ratio",
    ]
)

index = 0
stor = []
with open("../data/humanVirusSeq.fa") as d:
    lines = d.readlines()
    print(f"Number of Rows: {len(lines)}")
    for line in lines:
        if index % 2 == 0:
            stor = line.strip().replace("\n", "").replace(">", "").split("|")
            index += 1
        else:
            genome_data.loc[len(genome_data.index)] = [
                stor[0],
                stor[1],
                stor[2],
                line.strip(),
                None,
                0,
            ]
            index += 1

display(genome_data.head())

print(f"Length of Genome of Row 0: {len(genome_data.iloc[0]['Genome'].strip())}")
```

##### Virus Data (not DNA)

```python
virus_data = pd.read_excel("../data/Woolhouse_and_Brierley_RNA_virus_database.xlsx")
virus_data.head()
```

```python
i, j, compValues = 0, 0, []

for j in range(1620):
    while True:
        compValues.append(
            df.SequenceMatcher(
                None, virus_data.iloc[i, 0], genome_data.iloc[j, 2]
            ).ratio()
        )
        i += 1
        if i == 213:
            i = 0
            compValues.append(1.0)
            norm_compValues = [
                (float(i) - min(compValues)) / (max(compValues) - min(compValues))
                for i in compValues
            ]
            for k in range(213):
                if norm_compValues[k] > 0.75:
                    genome_data.loc[j, "Estimated Transmission Level"] = (
                        virus_data.iloc[k, 20]
                    )
                    break
            break
    compValues = []
    j += 1
    # print(j)

display(genome_data)
display(genome_data.loc[52])
```

#### Dropping and other Stuff

```python
deleteList = []

for i in range(len(genome_data)):
    if genome_data.loc[i, "Estimated Transmission Level"] == None:
        deleteList.append(i)

print(f"The rows deleted: {deleteList}")

genome_data = genome_data.drop(deleteList)
genome_data = genome_data.reset_index(drop=True)
```

#### Reading out the code

```python
genome_data = genome_data.drop(columns="temp ratio")
```
